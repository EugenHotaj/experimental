In this exercise, we are given a stimulus and the responses of 4 neurons on that stimulus.

```python
%matplotlib inline

import pickle

import matplotlib.pyplot as plt
import numpy as np

with open('tuning.pickle', 'rb') as f:
    data = pickle.load(f)
```

The below are plots of the average firing rates as a function of the stimulus for each neuron. 

```python
neuron_names = ['neuron{}'.format(i) for i in range(1,5)]
stim = data['stim']
avgs = [np.mean(np.array(data[neuron]), axis=0) for neuron in neuron_names]
for i in range(4):
    plt.xlabel('stimulus')
    plt.ylabel('neuron firing rate')
    plt.title(neuron_names[i])
    plt.plot(stim, avgs[i])
    plt.show()
```

```python
probs = [val/sum(avgs[0]) for val in avgs[0]]
mu = sum(stim*probs)
var = sum((stim-mu)**2*probs)
print(mu/10)
print(var/100)
print((mu/10)/(var/100))
```
